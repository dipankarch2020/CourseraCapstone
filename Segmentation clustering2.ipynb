## Step 1 : Collecting Data

```python
import pandas as pd
from bs4 import BeautifulSoup
from urllib.request import urlopen

import 

url="https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
html=urlopen(url)
soup = BeautifulSoup(html, 'lxml')
type(soup)

```

## Step 2 : Structuring and refining Data 

```python
rows=soup.find_all('tr')
print()

df=pd.DataFrame()
data=[]
for row in rows:
    row_str=str(row.text)
    lst=row_str.split("\n")
    #print(lst)
    #print(lst[1])
    row_list=[]
    for x in lst:
        if x !='':
            row_list.append(x)
    data.append(row_list)
    #print(row_list)
#print(data)
df=pd.DataFrame(data)
#print(df)
df_refined=pd.DataFrame([])
df_refined=df[:180]

#df_refined.replace(to_replace="Not Assigned", value="NA",inplace=True)
#df_refined.replace(to_replace="None", value="NA",inplace=True)
df_refined.dropna(axis=1,inplace=True,how='all')
df_refined.dropna(axis=0,inplace=True)

header=df_refined.iloc[0]
df_refined=df_refined[1:]
df_refined.columns=header
#df_refined
#type(df_refined)
df_refined.reset_index(drop=True,inplace=True)
df_refined

  

```

## Step 3 : Adding longitude and latitude

```python
long_lat_all=pd.read_csv("Geospatial_Coordinates.csv")
postal_codes=df_refined["Postal Code"]
#print(long_lat_all["Postal Code"][0])

long_lat_list=[]
long_lat=pd.DataFrame([['0','0']])
for p_code in postal_codes:
        for n in range(0,103):
            if p_code == long_lat_all["Postal Code"][n]:
                #print(p_code,long_lat_all["Postal Code"][n])
                #lnlat=pd.DataFrame([long_lat_all["Latitude"][n],long_lat_all["Longitude"][n]])
                lst=[float(long_lat_all["Latitude"][n]),float(long_lat_all["Longitude"][n])]
                long_lat_list=long_lat_list+[lst]
                #lnlat=lnlat.transpose()
                #long_lat.append(lnlat)
                #print(long_lat_all["Latitude"][n],long_lat_all["Longitude"][n])
#print(long_lat_list)
long_lat=pd.DataFrame(long_lat_list)


long_lat.columns=['Latitude','Longitude']
print(long_lat)

df_final=pd.concat([df_refined,long_lat],axis=1)
df_final[:20]
```

```python
import geopy.geocoders
from geopy.geocoders import Nominatim
#from nominatim import Nominatim
address = 'Toronto,Ontario'

geolocator = Nominatim(user_agent="tto_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto City are {}, {}.'.format(latitude, longitude))
```

```python
import folium

Toronto_map = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df_final['Latitude'], df_final['Longitude'], df_final['Borough'], df_final['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(Toronto_map)  
    

display(Toronto_map)
```
